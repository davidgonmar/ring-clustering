```python
import matplotlib.pyplot as plt
import numpy as np

# Define parameters
radius = 3
angle_deg = 45
angle_rad = np.radians(angle_deg)

# Calculate points
V_i = np.array([0, 0])
A = np.array([radius * np.cos(angle_rad), 0])
X_j = np.array([-1, -1])
r_d_ratio = radius / np.linalg.norm(X_j - V_i)
V_prime_i = r_d_ratio * X_j + (1 - r_d_ratio) * V_i

# Plotting the figure
fig, ax = plt.subplots()
circle = plt.Circle(V_i, radius, color='black', fill=False)
ax.add_artist(circle)

# Draw the triangle
ax.plot([V_i[0], A[0]], [V_i[1], A[1]], 'k-')
ax.plot([A[0], V_prime_i[0]], [A[1], V_prime_i[1]], 'k-')
ax.plot([V_prime_i[0], V_i[0]], [V_prime_i[1], V_i[1]], 'k-')

# Draw the dotted line
ax.plot([V_prime_i[0], X_j[0]], [V_prime_i[1], X_j[1]], 'k:', linewidth=1)

# Draw points
ax.plot(V_i[0], V_i[1], 'ko')
ax.plot(V_prime_i[0], V_prime_i[1], 'ko')
ax.plot(X_j[0], X_j[1], 'ko')

# Annotate points
ax.annotate(r'$V_i$', xy=V_i, xytext=(-0.5, -0.5), textcoords='offset points')
ax.annotate(r"$V'_i$", xy=V_prime_i, xytext=(5, 5), textcoords='offset points')
ax.annotate(r'$X_j$', xy=X_j, xytext=(5, -5), textcoords='offset points')

# Label the hypotenuse
mid_point = (V_i + V_prime_i) / 2
ax.annotate(r'$r_i$', xy=mid_point, xytext=(5, 5), textcoords='offset points')

# Explanation below the circle
ax.annotate(r"$V'_i = \frac{r_i}{d_{ij}}X_j + (1 - \frac{r_i}{d_{ij}})V_i$",
            xy=(0, -radius - 0.5), xycoords='data', ha='center', va='top')

# Set limits and aspect ratio
ax.set_xlim(-radius-1, radius+1)
ax.set_ylim(-radius-1, radius+1)
ax.set_aspect('equal')

# Remove axes
ax.axis('off')

# Show plot
plt.title('Illustration of the update of the cluster centers equation when X_j is inside the circle')
plt.show()
```
